## Set up SIMECO

Clone the repository and install dependencies. It may take 1-2 minutes.

```python
%%bash
git clone https://github.com/submission13448/simeco.git
# install pointnet2_ops_lib
cd simeco/extensions/pointnet2_ops_lib && python setup.py install
# install required PyPI packages
cd ../../build && pip install -r requirements.txt
# copy visual.py to root directory
cd ../..  && cp simeco/utils/visual.py visual.py
```

##  Quick examples

**Example without transformation**

You can interactively explore the model by dragging it.

```python
from visual import *
!python simeco/inference.py --pc_file simeco/data/pc/2a05d684eeb9c1cfae2ca9bb680dd18b.npy
visualize_point_cloud(
    "Incomplete point cloud", "simeco/data/pc/2a05d684eeb9c1cfae2ca9bb680dd18b.npy",
    "Predicted point cloud", "simeco/data/result/2a05d684eeb9c1cfae2ca9bb680dd18b.npy",
    eye = dict(x=0, y=0.8, z=0),
)
```
